```python
# Generated from ChatGPT: A simple plot of Sine Wave
import numpy as np
import plotly.graph_objects as go

x = np.linspace(0, 2*np.pi, 200)
y = np.sin(x)

fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y, mode='lines'))

fig.update_layout(title="Interactive Sine Wave")

fig.write_html("sine_wave.html")
```

```python
# Generated from ChatGPT: A time-slider based plot of the Wave Equation:
import numpy as np
import plotly.graph_objects as go

x = np.linspace(0, 1, 200)
t_vals = np.linspace(0, 2, 50)

fig = go.Figure()

for t in t_vals:
    y = np.sin(np.pi * x) * np.cos(np.pi * t)
    fig.add_trace(go.Scatter(x=x, y=y, visible=False))

fig.data[0].visible = True

steps = []
for i in range(len(t_vals)):
    step = dict(
        method="update",
        args=[{"visible": [j == i for j in range(len(t_vals))]}],
        label=f"{t_vals[i]:.2f}"
    )
    steps.append(step)

sliders = [dict(active=0, steps=steps)]

fig.update_layout(sliders=sliders, title="Wave Equation Evolution")

fig.write_html("wave.html")
```
